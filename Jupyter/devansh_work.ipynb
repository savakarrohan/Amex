```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import sys
import os
```

```python
df_train = pd.read_parquet('./train_data.parquet', engine='pyarrow')
```

```python
df_train
```

```python
df_train['missing_count'] = df_train.apply(lambda row: row.isnull().sum(), axis=1)
df_train['missing_count']
```

```python
df_train['missing_count']
```

```python
l = []
for x in df_train['missing_count']:
    l.append(x)
l = np.array(l)
print(l)
l.sort()
print(l, len(l))
ninety_percentile = np.percentile(l, 90)
ninety_percentile
```

```python
new_df = df_train[df_train['missing_count'] >= ninety_percentile]
new_df
```

```python
f_columns = [f'f{i}' for i in range(1, 367)]
# Row-wise median (ignoring NaN or None)
new_df[f_columns] = new_df[f_columns].apply(pd.to_numeric, errors='coerce')
new_df[f_columns] = new_df[f_columns].apply(lambda row: row.fillna(row.median()), axis=1)
new_df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_df = new_df[f_columns].apply(lambda row: scaler.fit_transform(row.values.reshape(-1, 1)).flatten(), axis=1, result_type='expand')
scaled_df.columns = new_df[f_columns].columns 
scaled_df
```

```python
y = new_df['y']
y = np.array(y)
x = np.array(scaled_df)
x.shape, y.shape
```

```python
y = [0 if i == '0' else 1 for i in y]
```

# Evaluation on Train Dataset

```python
from sklearn.model_selection import train_test_split
import xgboost as xgb
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
xgb_model = xgb.XGBClassifier(objective="binary:logistic", random_state=42)
xgb_model.fit(x_train, y_train)
y_pred = xgb_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score
print(f"Accuracy: ", accuracy_score(y_test, y_pred))
print(f"Confusion Matrix: ", confusion_matrix(y_test, y_pred))
print(f"Precision Score: ", precision_score(y_test, y_pred))
print("Classification Report:", classification_report(y_test, y_pred))
```
